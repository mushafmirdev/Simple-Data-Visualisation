# Data visualization

Histogram

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('Petal Length vs Petal Width by Species')
plt.show()
```

```python
#Feature Petal Length
plt.hist(df['PetalLengthCm'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Feature SepalLengthCm
plt.hist(df['SepalLengthCm'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

#Feature Species
plt.hist(df['Species'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Specie Type')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Box Plot for Iris Dataset Features')
plt.xticks(rotation=45)
plt.show()
```
